```python
#part 1

class Mortal_K_Game:
    def __init__(self, init_contestant):
        self.init_contestant = init_contestant
        self.maxim_depth = 5
        self.branchingF = 2
        self.utility_val = [-1, 1]
        self.times_played_so_far = 0
        self.victories_by_scorpion = 0
        self.subz_wins = 0
        self.round_won_by = []

    def game_reenactment(self):
        ongoing_contestant = self.init_contestant
        maxim_rounds = 3
        while self.times_played_so_far < maxim_rounds:
            self.times_played_so_far += 1
            champ = self.alpha_beta_prun(0, ongoing_contestant, float('-inf'), float('inf'), True)

            if champ == -1:
                self.round_won_by.append("Scorpion")
                self.victories_by_scorpion += 1
            else:
                self.round_won_by.append("Sub-Zero")
                self.subz_wins += 1

            if self.victories_by_scorpion == 2:
                print("Scorpion wins the game by winning 2 rounds!")
                break
            elif self.subz_wins == 2:
                print("Sub-Zero wins the game by winning 2 rounds!")
                break

            ongoing_contestant = 1 - ongoing_contestant

        if self.victories_by_scorpion > self.subz_wins:
          victory_by = "Scorpion"
          return victory_by
        else:
          victory_by = "Sub-Zero"
          return victory_by

    def alpha_beta_prun(self, deepness, player, alpha_val, beta_val, maximizing_contestant):
        if deepness == self.maxim_depth:
            return self.utility_val[player]

        if maximizing_contestant:
            maxim_eval = float('-inf')
            for x in range(self.branchingF):
                eval = self.alpha_beta_prun(deepness + 1, player, alpha_val, beta_val, False)
                maxim_eval = max(maxim_eval, eval)
                alpha_val = max(alpha_val, eval)
                if beta_val <= alpha_val:
                    break
            return maxim_eval

        else:
            minim_eval = float('inf')
            for y in range(self.branchingF):
                eval = self.alpha_beta_prun(deepness + 1, 1 - player, alpha_val, beta_val, True)
                minim_eval = min(minim_eval, eval)
                beta_val = min(beta_val, eval)
                if beta_val <= alpha_val:
                    break
            return minim_eval

init_contestant = int(input("who is the init player? (0 for Scorpion, 1 for Sub-Zero): "))
game = Mortal_K_Game(init_contestant)

game_won_by = game.game_reenactment()

print(f"Game champ: {game_won_by}")
print(f"Total Rounds Played: {game.times_played_so_far}")
for count, champ in enumerate(game.round_won_by, 1):
    print(f"Winner of the Round {count}: {champ}")
```

```python
#part 2

def pacman_game(c):
    nodes_at_leaf_vals = [3, 6, 2, 3, 7, 1, 2, 0]
    def mini_max(deepness, idx, is_it_the_maximizing_player, given_vals, alpha, beta):
        if deepness == 3:
            return given_vals[idx]
        if is_it_the_maximizing_player:
            max_placeholder = float('-inf')
            for times in range(2):
                evaluated_eval = mini_max(deepness + 1, idx * 2 + times, False, given_vals, alpha, beta)
                max_placeholder = max(max_placeholder, evaluated_eval)
                alpha = max(alpha, evaluated_eval)
                if beta <= alpha:
                    break
            return max_placeholder
        else:
            minim_placeholder = float('inf')
            for times in range(2):
                evaluated_eval = mini_max(deepness + 1, idx * 2 + times, True, given_vals, alpha, beta)
                minim_placeholder = min(minim_placeholder, evaluated_eval)
                beta = min(beta, evaluated_eval)
                if beta <= alpha:
                    break
            return minim_placeholder

    mini_max_val = mini_max(0, 0, True, nodes_at_leaf_vals, float('-inf'), float('inf'))

    sub_max_fromleft = max(nodes_at_leaf_vals[0:4])
    sub_max_fromright = max(nodes_at_leaf_vals[4:8])
    when_utilized_left_side_magic = sub_max_fromleft - c
    when_utilized_right_side_magic = sub_max_fromright - c

    if when_utilized_left_side_magic > when_utilized_right_side_magic and when_utilized_left_side_magic > mini_max_val:
        print(f"The new value of minimax is {when_utilized_left_side_magic}. Pacman goes left and uses dark magic")
    elif when_utilized_right_side_magic > when_utilized_left_side_magic and when_utilized_right_side_magic > mini_max_val:
        print(f"The new value of minimax is {when_utilized_right_side_magic}. Pacman goes right and uses dark magic")
    else:
        print(f"The minimax value is {mini_max_val}. Pacman does not use dark magic")

pacman_game(2)
pacman_game(5)
```
